## Most available room of all time 

```python
most_available_place=df.nlargest(10,'availability_365')
most_available_place.head(5)
```

```python
# Average price of most available place
price_avrg=most_available_place.price.mean()
print('Average price of most available place: {}'.format(price_avrg))
```

## Average price based on minimum_nights of stay in descending order  

```python
min_nights_stay = df.groupby(['minimum_nights','neighbourhood_group','neighbourhood'],dropna=True)['price'].mean().reset_index()
```

```python
min_nights_stay.sort_values('minimum_nights',ascending=False,inplace=True)
min_nights_stay
```

```python
plt.figure(figsize=(11,7))
plt.scatter(min_nights_stay['minimum_nights'],min_nights_stay['price'])
plt.ylim(0, 600)
plt.xlabel('Minimum_nights')
plt.ylabel('Price')
```

## Average price per night for each neighbourhood along with room_type

```python
avg_price_per_night_nbd = df.groupby(['neighbourhood_group','room_type'])['price'].mean().reset_index()
avg_price_per_night_nbd
```

```python
sns.barplot(x=avg_price_per_night_nbd['neighbourhood_group'],hue=avg_price_per_night_nbd['room_type'],y=avg_price_per_night_nbd['price'])
```

## Count the number of rooms in every neighbourhood group

```python
count_of_room = df.groupby(['neighbourhood_group']).room_type.count().sort_values().reset_index().rename(columns={'room_type':'Count of room'})
count_of_room
```

```python
plt.figure(figsize=(11,7))
ax = sns.barplot(data = count_of_room.sort_values(by ='Count of room',ascending = False), x ='neighbourhood_group',y='Count of room')
plt.title('Count of rooms vs Neighbourhood Group',fontsize=17)
plt.suptitle('2019 NYC Airbnb Data',fontsize=14)
```
